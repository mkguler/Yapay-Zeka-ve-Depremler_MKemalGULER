```python
# gerekli kütüphanelerin aktarımı
import csv
from datetime import datetime

#  Menu.txt dosyasını Google Driveda oluşturarak ID numberını belirtip projenin içine çağırıyoruz
!gdown --id 1_RT9okbQ6W5l9zsToVzR4NvV3n4fjywk


# Öncelikle bir “Pizza” superclassı oluşturuyoruz
class Pizza:
    fiyat = 0.0
    description = ""

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.fiyat

# Sonra bir “Pizza” subclassı oluşturup pizza çeşitlerini tanımlıyoruz
class Klasik(Pizza):
    def __init__(self):
        self.description = "Klasik Pizza"
        self.fiyat = 90.00

class Margarita(Pizza):
    def __init__(self):
        self.description = "Margarita Pizza"
        self.fiyat = 80.99

class TurkPizza(Pizza):
    def __init__(self):
        self.description = "Turk Pizza"
        self.fiyat = 100.99

class SadePizza(Pizza):
    def __init__(self):
        self.description = "Sade Pizza"
        self.fiyat = 70.99

# Bu aşamada soslar için bir “Decorator” superclassı tanumlıyoruz
class Decorator(Pizza):
    component = None

    def __init__(self, component):
        self.component = component

    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self)

    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self)

# tekrardan bir “Decorator” subclassı oluşturarak sosları ve fiyatlarını tanımlıyoruz
class Zeytin(Decorator):
    def __init__(self, component):
        Decorator.__init__(self, component)
        self.fiyat = 10.00
        self.description = "Zeytin"

class Mantar(Decorator):
    def __init__(self, component):
        Decorator.__init__(self, component)
        self.fiyat = 11.50
        self.description = "Mantar"

class KeciPeyniri(Decorator):
    def __init__(self, component):
        Decorator.__init__(self, component)
        self.fiyat = 20.00
        self.description = "KeciPeyniri"

class Et(Decorator):
    def __init__(self, component):
        Decorator.__init__(self, component)
        self.fiyat = 22.50
        self.description = "Et"

class Sogan(Decorator):
    def __init__(self, component):
        Decorator.__init__(self, component)
        self.fiyat = 10.25
        self.description = "Sogan"

class Misir(Decorator):
    def __init__(self, component):
        Decorator.__init__(self, component)
        self.fiyat = 8.00
        self.description = "Misir"

#  main function yaratarak kullanıcı arayüzü oluşturuyoruz
def main():
    print("Pizza Siparis Sistemine Hosgeldiniz!")
    print("\nMenu:")
    with open('Menu.txt') as menu_file:
        menu_text = menu_file.read()
        print(menu_text)

    pizza_choice = int(input("\nLutfen Bir Pizza Tabanı Seciniz: "))
    if pizza_choice == 1:
        pizza = Klasik()
    elif pizza_choice == 2:
        pizza = Margarita()
    elif pizza_choice == 3:
        pizza = TurkPizza()
    elif pizza_choice == 4:
        pizza = SadePizza()
    else:
        print("Gecersiz secim.")
        return

    sauce_choice = int(input("ve sececeginiz sos: "))
    if sauce_choice == 11:
        pizza = Zeytin(pizza)
    elif sauce_choice == 12:
        pizza = Mantar(pizza)
    elif sauce_choice == 13:
        pizza = KeciPeyniri(pizza)
    elif sauce_choice == 14:
        pizza = Et(pizza)
    elif sauce_choice == 15:
        pizza = Sogan(pizza)
    elif sauce_choice == 16:
        pizza = Misir(pizza)
    else:
        print("Gecersiz Sos Secimi.")
        return

    total_cost = pizza.get_cost()
    print("\nSiparisiniz", pizza.get_description(), "ile", pizza.component.get_description())
    print("Toplam Odeme: ₺", total_cost)

    name = input("\nAdiniz Soyadiniz: ")
    id_number = input("TC Kimlik Numaranizi Giriniz: ")
    credit_card_number = input("Kredi Karti Numaranizi Giriniz: ")
    credit_card_password = input("Kredi Karti Sifrenizi Giriniz: ")

    with open('Orders_Database.csv', mode='a', newline='') as orders_file:
        writer = csv.writer(orders_file)
        now = datetime.now()
        date_time = now.strftime("%m/%d/%Y %H:%M:%S")
        writer.writerow([name, id_number, credit_card_number, credit_card_password, pizza.get_description(),
                         pizza.component.get_description(), total_cost, date_time])

    print("\nSiparisiniz icin Tesekkur Ederiz!")

if __name__ == '__main__':
    main()
```
